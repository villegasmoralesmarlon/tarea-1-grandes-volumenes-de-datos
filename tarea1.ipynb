```python
import pandas as pd
df = pd.read_csv("transacciones_simuladas 2.csv")
print(df.head())
```

```python
#1.Si el monto  > 45000
df[df["monto"]>45000]
```

```python
#2.Si tipo_transaccion es retiro y ciudad es extranjera y el monto > 25000:
df[(df["tipo_transaccion"] == "retiro") & (~df["ciudad"].isin(["Mexico","CDMX","Guadalajara","Monterrey","Puebla","Cancun","Merida","Tijuana","Queretaro","Chihuahua","Hermosillo","Culiacan","Toluca","Leon","Morelia","Villahermosa","Tuxtla","Veracruz","Oaxaca","Zacatecas","Durango","Saltillo","Aguascalientes","Campeche","Colima","Cuernavaca","Chetumal","Tlaxcala","Juarez","Torreon"])) & (df["monto"] > 25000)]
```

```python
#3.tipo_transaccion es transferencia y el monto > 30000 o se realizó desde una ciudad fuera de México
df[(df["tipo_transaccion"] == "transferencia") & ((df["monto"] > 30000) | (~df["ciudad"].isin(["Mexico","CDMX","Guadalajara","Monterrey","Puebla","Cancun","Merida","Tijuana","Queretaro","Chihuahua","Hermosillo","Culiacan","Toluca","Leon","Morelia","Villahermosa","Tuxtla","Veracruz","Oaxaca","Zacatecas","Durango","Saltillo","Aguascalientes","Campeche","Colima","Cuernavaca","Chetumal","Tlaxcala","Juarez","Torreon"])))]
```

```python
#4.Si tipo_transaccion == 'compra' y 'categoria_comercio' == 'viajes' and 'monto' > 35000:
df[(df["tipo_transaccion"] == "compra") & (df["categoria_comercio"] == "viajes") & (df["monto"] > 35000)]
```
